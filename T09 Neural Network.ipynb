## i) Neuron (Artificial Neuron)

An artificial neuron, also known as a node or unit, is the basic building block of an ANN. It mimics the behavior of a biological neuron to a certain extent. A list of breakdown components is as follows:

1.  **Inputs ($x_1, x_2, ..., x_n$):** These are the signals or data received by the neuron from the external environment or from other neurons in the network. Each input is typically a numerical value.

2.  **Weights ($w_1, w_2, ..., w_n$):** Each input connection has an associated weight. These weights represent the strength or importance of the corresponding input. During the learning process, the network adjusts these weights to optimize its performance.

3.  **Weighted Sum:** The neuron calculates a weighted sum of its inputs. Each input is multiplied by its corresponding weight, and these products are summed together.

4.  **Bias ($b$):** A bias term is an additional constant input to the neuron, often with a weight of 1. It allows the neuron to be activated even when all inputs are zero, providing an extra degree of freedom in the model. The weighted sum is added to the bias.

5.  **Activation Function:** The result of the weighted sum (plus bias), denoted as $z$, is passed through an activation function. This function introduces non-linearity into the neuron's output. Non-linearity is crucial for ANNs to learn complex patterns in data. Common activation functions include sigmoid, ReLU (Rectified Linear Unit), tanh (hyperbolic tangent), etc.

6.  **Output ($a$):** The output of the activation function, denoted as $a$, is the final output of the neuron. This output can then be passed as an input to other neurons in the network.



```python
import numpy as np

def sigmoid(x):
    """Sigmoid activation function."""
    return 1 / (1 + np.exp(-x))

# Inputs
inputs = np.array([1.0, 0.5, -2.0])

# Weights
weights = np.array([0.2, -0.3, 0.8])

# Bias
bias = 0.1

# Calculate the weighted sum
weighted_sum = np.dot(inputs, weights) + bias

# Apply the activation function
output = sigmoid(weighted_sum)

print(f"Inputs: {inputs}")
print(f"Weights: {weights}")
print(f"Bias: {bias}")
print(f"Weighted Sum (z): {weighted_sum:.4f}")
print(f"Neuron Output (a): {output:.4f}")
```

## ii) Perceptron

A perceptron is one of the earliest and simplest types of artificial neural networks. It is essentially a single-layer neural network consisting of a single neuron with a **step activation function** (or a similar threshold-based function).

1.  **Structure:** A perceptron takes multiple inputs, applies weights and a bias (implicitly or explicitly), calculates the weighted sum, and then passes this sum through a step function.

2.  **Step Activation Function:** The key characteristic of a traditional perceptron is its step activation function. This function outputs a binary value (0 or 1) based on whether the weighted sum (plus bias) exceeds a certain threshold (often 0).

3.  **Linear Separability:** Perceptrons are primarily used for binary classification problems where the classes are linearly separable. They can learn a linear decision boundary to separate the two classes.

4.  **Limitations:** Perceptrons have significant limitations. They cannot learn complex non-linear patterns, such as the XOR problem. Multi-layer perceptrons (feedforward neural networks with multiple layers and non-linear activation functions) were developed to overcome these limitations.


```python
import numpy as np

def step_function(x, threshold=0):
    """Step activation function."""
    return 1 if x >= threshold else 0

# Inputs
inputs = np.array([1.0, 0.5, -2.0])

# Weights
weights = np.array([0.2, -0.3, 0.8])

# Bias
bias = 0.1

# Threshold
threshold = 0.5

# Calculate the weighted sum
weighted_sum = np.dot(inputs, weights) + bias

# Apply the step function
output = step_function(weighted_sum, threshold)

print(f"Inputs: {inputs}")
print(f"Weights: {weights}")
print(f"Bias: {bias}")
print(f"Weighted Sum (z): {weighted_sum:.4f}")
print(f"Perceptron Output: {output}")
```

## iii) Activation Function

An activation function is a crucial component of an artificial neuron. It determines the output of the neuron based on its input (the weighted sum plus bias). The primary role of activation functions is to introduce **non-linearity** into the neural network. Without non-linear activation functions, a multi-layer neural network would essentially behave like a single-layer linear model, limiting its ability to learn complex relationships in data.

**Common Activation Functions and their Python Implementations:**

1.  **Sigmoid (Logistic):**
    * Formula: $\sigma(z) = \frac{1}{1 + e^{-z}}$
    * Output Range: (0, 1)
    * Use Cases: Binary classification (output as a probability).









```python
    import numpy as np
    import matplotlib.pyplot as plt

    def sigmoid(z):
        return 1 / (1 + np.exp(-z))

    z = np.linspace(-5, 5, 100)
    a = sigmoid(z)

    plt.plot(z, a)
    plt.xlabel('Input (z)')
    plt.ylabel('Sigmoid Output (a)')
    plt.title('Sigmoid Activation Function')
    plt.grid(True)
    plt.show()
```

2.  **ReLU (Rectified Linear Unit):**
    * Formula: $ReLU(z) = \max(0, z)$
    * Output Range: $[0, \infty)$
    * Use Cases: Most common activation function for hidden layers in deep learning due to its simplicity and efficiency.

```python
    import numpy as np
    import matplotlib.pyplot as plt

    def relu(z):
        return np.maximum(0, z)

    z = np.linspace(-5, 5, 100)
    a = relu(z)

    plt.plot(z, a)
    plt.xlabel('Input (z)')
    plt.ylabel('ReLU Output (a)')
    plt.title('ReLU Activation Function')
    plt.grid(True)
    plt.show()
```

3.  **Tanh (Hyperbolic Tangent):**
    * Formula: $\tanh(z) = \frac{e^{z} - e^{-z}}{e^{z} + e^{-z}}$
    * Output Range: (-1, 1)
    * Use Cases: Similar to sigmoid but with a centered output, which can sometimes lead to faster training.

```python
    import numpy as np
    import matplotlib.pyplot as plt

    def tanh(z):
        return np.tanh(z)

    z = np.linspace(-5, 5, 100)
    a = tanh(z)

    plt.plot(z, a)
    plt.xlabel('Input (z)')
    plt.ylabel('Tanh Output (a)')
    plt.title('Tanh Activation Function')
    plt.grid(True)
    plt.show()
```

There are many other activation functions like Softmax (for multi-class classification output), Leaky ReLU, ELU, etc., each with its own characteristics and use cases.

## iv) Backpropagation

Backpropagation is a supervised learning algorithm used to train artificial neural networks. It works by calculating the gradient of the loss function with respect to the network's weights and biases and then using this gradient to update the weights and biases in the direction that minimizes the loss. It's the core mechanism that allows neural networks to learn from data.

Here's a step-by-step overview of the backpropagation process:

1.  **Forward Pass:**
    * Input data is fed into the network.
    * The input propagates through the network, layer by layer.
    * At each neuron, a weighted sum of inputs is calculated, and then an activation function is applied to produce the output of that neuron.
    * This process continues until the output layer produces a prediction.

2.  **Loss Calculation:**
    * The output of the network is compared to the actual target values using a loss function (e.g., Mean Squared Error for regression, Cross-Entropy for classification). The loss function quantifies the error between the prediction and the true value.

3.  **Backward Pass (Backpropagation of Error):**
    * The gradient of the loss function with respect to the output layer's parameters (weights and biases) is calculated.
    * This gradient is then propagated backward through the network, layer by layer, using the chain rule of calculus.
    * At each layer, the gradient of the loss function with respect to the weights and biases of that layer is calculated. This indicates how much a small change in each weight or bias would affect the overall loss.

4.  **Weight and Bias Update:**
    * Once the gradients are calculated, the weights and biases of the network are updated using an optimization algorithm (e.g., Gradient Descent, Adam, RMSprop).
    * The update rule typically involves subtracting a fraction of the gradient (scaled by a learning rate) from the current weight or bias. This moves the parameters in the direction that reduces the loss.

```python
# The code demonstrates the basic principles of backpropagation
# in a simple two-layer neural network.

import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return sigmoid(x) * (1 - sigmoid(x))

# Training data (simple example)
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([[0], [1], [1], [0]])  # XOR problem

# Network architecture
input_size = 2
hidden_size = 2
output_size = 1
learning_rate = 0.1
epochs = 10000

# Initialize weights and biases randomly
W1 = np.random.randn(input_size, hidden_size)
b1 = np.random.randn(hidden_size)
W2 = np.random.randn(hidden_size, output_size)
b2 = np.random.randn(output_size)

# Training loop
for epoch in range(epochs):
    # Forward pass
    z1 = np.dot(X, W1) + b1
    a1 = sigmoid(z1)
    z2 = np.dot(a1, W2) + b2
    a2 = sigmoid(z2)  # Output

    # Calculate the loss (Mean Squared Error)
    loss = np.mean((a2 - y)**2)

    # Backward pass (Backpropagation)
    # Output layer
    dz2 = (a2 - y) * sigmoid_derivative(z2)
    dW2 = np.dot(a1.T, dz2)
    db2 = np.sum(dz2, axis=0, keepdims=True)

    # Hidden layer
    da1 = np.dot(dz2, W2.T)
    dz1 = da1 * sigmoid_derivative(z1)
    dW1 = np.dot(X.T, dz1)
    db1 = np.sum(dz1, axis=0, keepdims=True)

    # Update weights and biases
    W2 -= learning_rate * dW2
    b2 -= learning_rate * db2.squeeze()
    W1 -= learning_rate * dW1
    b1 -= learning_rate * db1.squeeze()

    if epoch % 1000 == 0:
        print(f"Epoch {epoch}, Loss: {loss:.4f}")

# Test the trained network
print("\nPredictions after training:")
z1_test = np.dot(X, W1) + b1
a1_test = sigmoid(z1_test)
z2_test = np.dot(a1_test, W2) + b2
a2_test = sigmoid(z2_test)
print(a2_test)
```

Steps:
1.  **Forward Pass:** The code calculates the outputs of the hidden layer (`a1`) and the output layer (`a2`) by performing weighted sums and applying the sigmoid activation function.
2.  **Loss Calculation:** The Mean Squared Error (MSE) between the predicted output (`a2`) and the true target (`y`) is calculated.
3.  **Backward Pass:**
    * The error at the output layer is calculated (`dz2`).
    * Gradients of the loss with respect to the output layer weights (`dW2`) and bias (`db2`) are computed.
    * The error is then backpropagated to the hidden layer (`da1`, `dz1`).
    * Gradients of the loss with respect to the hidden layer weights (`dW1`) and bias (`db1`) are computed.
4.  **Weight and Bias Update:** The weights and biases of both layers are updated using the calculated gradients and the learning rate. The learning rate controls the step size during the optimization process.
5.  **Training Loop:** The forward and backward passes are repeated for a specified number of epochs (iterations) to minimize the loss and improve the network's performance.

## Single-Layer Perceptron Network

```python
import numpy as np

class SingleLayerPerceptron:

    def __init__(self, num_inputs, activation_function='step'):
        self.num_inputs = num_inputs
        self.weights = np.random.rand(num_inputs)  # Initialize weights randomly
        self.bias = np.random.rand(1)[0]          # Initialize bias randomly
        self.activation_function_name = activation_function
        self.activation_function = self._get_activation_function(activation_function)
        self.derivative_activation = self._get_derivative_activation(activation_function)

    def _get_activation_function(self, name):
        if name == 'step':
            return lambda x: 1 if x >= 0 else 0
        elif name == 'sigmoid':
            return lambda x: 1 / (1 + np.exp(-x))
        elif name == 'tanh':
            return np.tanh
        elif name == 'relu':
            return lambda x: np.maximum(0, x)
        else:
            raise ValueError(f"Activation function '{name}' not supported.")

    def _get_derivative_activation(self, name):
        if name == 'step':
            return lambda x: 0  # Derivative is 0 almost everywhere
        elif name == 'sigmoid':
            return lambda x: self.activation_function(x) * (1 - self.activation_function(x))
        elif name == 'tanh':
            return lambda x: 1 - np.tanh(x)**2
        elif name == 'relu':
            return lambda x: 1 if x > 0 else 0
        else:
            raise ValueError(f"Derivative for activation function '{name}' not supported.")

    def predict(self, inputs):
        if inputs.shape[0] != self.num_inputs:
            raise ValueError(f"Input size mismatch. Expected {self.num_inputs}, got {inputs.shape[0]}.")

        # Summation of weighted inputs and bias
        weighted_sum = np.dot(inputs, self.weights) + self.bias

        # Apply the activation function
        output = self.activation_function(weighted_sum)
        return output

# Define the number of input features
num_inputs = 3

# Create a single-layer perceptron with sigmoid activation
perceptron_sigmoid = SingleLayerPerceptron(num_inputs=num_inputs, activation_function='sigmoid')
print("Perceptron (Sigmoid):")
print(f"  Initial Weights: {perceptron_sigmoid.weights}")
print(f"  Initial Bias: {perceptron_sigmoid.bias}")

# Example input
input_data = np.array([0.8, -0.5, 1.2])
prediction_sigmoid = perceptron_sigmoid.predict(input_data)
print(f"  Prediction for input {input_data}: {prediction_sigmoid:.4f}")

# Create a single-layer perceptron with step activation
perceptron_step = SingleLayerPerceptron(num_inputs=num_inputs, activation_function='step')
print("\nPerceptron (Step):")
print(f"  Initial Weights: {perceptron_step.weights}")
print(f"  Initial Bias: {perceptron_step.bias}")
prediction_step = perceptron_step.predict(input_data)
print(f"  Prediction for input {input_data}: {prediction_step}")

# Create a single-layer perceptron with ReLU activation
perceptron_relu = SingleLayerPerceptron(num_inputs=num_inputs, activation_function='relu')
print("\nPerceptron (ReLU):")
print(f"  Initial Weights: {perceptron_relu.weights}")
print(f"  Initial Bias: {perceptron_relu.bias}")
prediction_relu = perceptron_relu.predict(input_data)
print(f"  Prediction for input {input_data}: {prediction_relu:.4f}")

# Create a single-layer perceptron with tanh activation
perceptron_tanh = SingleLayerPerceptron(num_inputs=num_inputs, activation_function='tanh')
print("\nPerceptron (Tanh):")
print(f"  Initial Weights: {perceptron_tanh.weights}")
print(f"  Initial Bias: {perceptron_tanh.bias}")
prediction_tanh = perceptron_tanh.predict(input_data)
print(f"  Prediction for input {input_data}: {prediction_tanh:.4f}")
```

The code depicts a perceptron with a single hidden layer, and demonstrates the fundamental principles of forward propagation and backpropagation for training a simple neural network.

* **Hidden Layer:** The network now has an intermediate layer (the hidden layer) between the input and output layers. This allows the network to learn more complex, non-linear relationships in the data compared to a single-layer perceptron.
* **Forward Pass:** Data flows from the input layer through the hidden layer to the output layer.
* **Backward Pass (Backpropagation):** The error is calculated at the output and propagated backward to adjust the weights in both the output and hidden layers.
* **Activation Function (Sigmoid):** The sigmoid function introduces non-linearity, enabling the network to learn complex patterns.
* **Weight and Bias Initialization:** Weights are initialized randomly (with small values), and biases are initialized to zero.
* **Learning Rate:** Controls the step size during weight updates.
* **Epochs:** The number of times the entire training dataset is passed through the network.
* **Loss Function (Binary Cross-Entropy):** Measures the error between the predicted and true outputs for binary classification.
* **Gradient Descent (Implicit):** The weights and biases are updated in the direction that reduces the loss.



## Multi-Layer Perceptron Network

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

class MultiLayerPerceptron:
    def __init__(self, input_size, hidden_size, output_size, learning_rate=0.1, epochs=1000):
        self.input_size = input_size
        self.hidden_size = hidden_size
        self.output_size = output_size
        self.learning_rate = learning_rate
        self.epochs = epochs

        # Initialize weights and biases for the hidden layer
        self.weights_hidden = np.random.randn(self.input_size, self.hidden_size) * 0.01
        self.bias_hidden = np.zeros((1, self.hidden_size))

        # Initialize weights and biases for the output layer
        self.weights_output = np.random.randn(self.hidden_size, self.output_size) * 0.01
        self.bias_output = np.zeros((1, self.output_size))

    def sigmoid(self, x):
        """Sigmoid activation function."""
        return 1 / (1 + np.exp(-x))

    def sigmoid_derivative(self, x):
        """Derivative of the sigmoid function."""
        return self.sigmoid(x) * (1 - self.sigmoid(x))

    def forward(self, X):
        """Forward pass through the network."""
        # Hidden layer
        self.hidden_layer_input = np.dot(X, self.weights_hidden) + self.bias_hidden
        self.hidden_layer_output = self.sigmoid(self.hidden_layer_input)

        # Output layer
        self.output_layer_input = np.dot(self.hidden_layer_output, self.weights_output) + self.bias_output
        self.output_layer_output = self.sigmoid(self.output_layer_input)

        return self.output_layer_output

    def backward(self, X, y, output):
        """Backward pass for backpropagation."""
        m = X.shape[0]  # Number of samples

        # Derivative of the loss function with respect to the output layer's input
        d_output_layer_input = (output - y) * self.sigmoid_derivative(self.output_layer_input)

        # Gradients for the output layer
        d_weights_output = np.dot(self.hidden_layer_output.T, d_output_layer_input) / m
        d_bias_output = np.sum(d_output_layer_input, axis=0, keepdims=True) / m

        # Derivative of the hidden layer's output with respect to its input
        d_hidden_layer_output = np.dot(d_output_layer_input, self.weights_output.T) * self.sigmoid_derivative(self.hidden_layer_input)

        # Gradients for the hidden layer
        d_weights_hidden = np.dot(X.T, d_hidden_layer_output) / m
        d_bias_hidden = np.sum(d_hidden_layer_output, axis=0, keepdims=True) / m

        return d_weights_hidden, d_bias_hidden, d_weights_output, d_bias_output

    def update_parameters(self, d_weights_hidden, d_bias_hidden, d_weights_output, d_bias_output):
        """Update weights and biases."""
        self.weights_hidden -= self.learning_rate * d_weights_hidden
        self.bias_hidden -= self.learning_rate * d_bias_hidden
        self.weights_output -= self.learning_rate * d_weights_output
        self.bias_output -= self.learning_rate * d_bias_output

    def train(self, X_train, y_train, X_val=None, y_val=None, verbose=True):
        """Trains the perceptron."""
        history = {'loss': [], 'val_loss': [], 'accuracy': [], 'val_accuracy': []}
        m = X_train.shape[0]

        for epoch in range(self.epochs):
            # Forward pass
            output = self.forward(X_train)

            # Calculate loss (Binary Cross-Entropy for binary classification)
            loss = -np.mean(y_train * np.log(output + 1e-8) + (1 - y_train) * np.log(1 - output + 1e-8))
            history['loss'].append(loss)

            # Calculate accuracy
            predictions = (output > 0.5).astype(int)
            accuracy = np.mean(predictions == y_train)
            history['accuracy'].append(accuracy)

            # Backward pass
            d_wh, d_bh, d_wo, d_bo = self.backward(X_train, y_train, output)

            # Update parameters
            self.update_parameters(d_wh, d_bh, d_wo, d_bo)

            # Validation (if validation data is provided)
            if X_val is not None and y_val is not None:
                val_output = self.forward(X_val)
                val_loss = -np.mean(y_val * np.log(val_output + 1e-8) + (1 - y_val) * np.log(1 - val_output + 1e-8))
                history['val_loss'].append(val_loss)
                val_predictions = (val_output > 0.5).astype(int)
                val_accuracy = np.mean(val_predictions == y_val)
                history['val_accuracy'].append(val_accuracy)

            if verbose and (epoch + 1) % (self.epochs // 10) == 0:
                print(f"Epoch {epoch + 1}/{self.epochs}, Loss: {loss:.4f}, Accuracy: {accuracy:.4f}", end="")
                if X_val is not None and y_val is not None:
                    print(f", Val Loss: {val_loss:.4f}, Val Accuracy: {val_accuracy:.4f}")
                else:
                    print()

        return history

    def predict(self, X):
        """Predicts the output for given input."""
        output = self.forward(X)
        return (output > 0.5).astype(int)

# Generate synthetic binary classification data
X, y = make_classification(n_samples=300, n_features=2, n_informative=2, n_redundant=0,
                           n_clusters_per_class=1, random_state=42)
y = y.reshape(-1, 1)

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Define the model
input_size = X_train.shape[1]  # Number of features
hidden_size = 4
output_size = 1  # Binary classification

model = MultiLayerPerceptron(input_size, hidden_size, output_size, learning_rate=0.1, epochs=5000)

# Train the model with validation
history = model.train(X_train, y_train, X_test, y_test, verbose=True)

# Evaluate the model on the test set
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"\nTest Accuracy: {accuracy:.4f}")

# Plotting the decision boundary (for 2D data)
if input_size == 2:
    plt.figure(figsize=(10, 6))
    plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train.flatten(), cmap=plt.cm.RdYlBu, label='Train Data')
    plt.scatter(X_test[:, 0], X_test[:, 1], marker='x', c=y_test.flatten(), cmap=plt.cm.RdYlBu, label='Test Data')

    # Create a meshgrid to plot decision boundary
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01),
                         np.arange(y_min, y_max, 0.01))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    plt.contourf(xx, yy, Z, cmap=plt.cm.RdYlBu, alpha=0.3)
    plt.title('Decision Boundary of the Perceptron with Hidden Layer')
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.legend()
    plt.show()

# Plot training history
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(history['loss'], label='Training Loss')
if 'val_loss' in history:
    plt.plot(history['val_loss'], label='Validation Loss')
plt.title('Loss over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history['accuracy'], label='Training Accuracy')
if 'val_accuracy' in history:
    plt.plot(history['val_accuracy'], label='Validation Accuracy')
plt.title('Accuracy over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.tight_layout()
plt.show()
```

## Roles of Activation Function

The primary role of an activation function within a neuron of a neural network is to introduce **non-linearity** to the model. Without non-linear activation functions, a neural network, no matter how many layers it has, would essentially behave like a single-layer linear model. This linear model would only be capable of learning linearly separable patterns in the data, severely limiting its ability to model complex real-world relationships.

Here's a breakdown of the key roles:

1.  **Introducing Non-Linearity:** This is the most critical role. Real-world data is often non-linear. Activation functions allow the network to learn these intricate patterns by mapping the linearly combined input to a non-linear output. This enables the network to approximate any complex function (Universal Approximation Theorem).

2.  **Decision Making (Thresholding):** Some activation functions (like the step function, though less common in deep networks) act as a threshold, determining whether a neuron should "fire" or not based on its input. This mimics the behavior of biological neurons.

3.  **Output Range Control:** Activation functions often squash the output of a neuron into a specific range (e.g., sigmoid to [0, 1], tanh to [-1, 1]). This can be useful for interpreting the output as probabilities (sigmoid) or for stabilizing the network's behavior.

4.  **Gradient Flow During Backpropagation:** The derivative of the activation function plays a crucial role during backpropagation. The gradients of the loss function are propagated backward through the network, and the derivative of the activation function at each neuron scales these gradients. The choice of activation function can significantly impact the efficiency and stability of the training process (e.g., vanishing/exploding gradients).

```python
import numpy as np
import matplotlib.pyplot as plt

# Input values
z = np.linspace(-5, 5, 100)

# 1. Sigmoid (Logistic) Activation
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

a_sigmoid = sigmoid(z)

plt.figure(figsize=(12, 6))
plt.subplot(2, 2, 1)
plt.plot(z, a_sigmoid)
plt.title('Sigmoid Activation')
plt.xlabel('Input (z)')
plt.ylabel('Output (a)')
plt.grid(True)

# 2. ReLU (Rectified Linear Unit) Activation
def relu(x):
    return np.maximum(0, x)

a_relu = relu(z)

plt.subplot(2, 2, 2)
plt.plot(z, a_relu)
plt.title('ReLU Activation')
plt.xlabel('Input (z)')
plt.ylabel('Output (a)')
plt.grid(True)

# 3. Tanh (Hyperbolic Tangent) Activation
def tanh(x):
    return np.tanh(x)

a_tanh = tanh(z)

plt.subplot(2, 2, 3)
plt.plot(z, a_tanh)
plt.title('Tanh Activation')
plt.xlabel('Input (z)')
plt.ylabel('Output (a)')
plt.grid(True)

# 4. Step Function (Illustrative - Less common in deep networks)
def step_function(x):
    return np.array([1 if val >= 0 else 0 for val in x])

a_step = step_function(z)

plt.subplot(2, 2, 4)
plt.plot(z, a_step)
plt.title('Step Function Activation')
plt.xlabel('Input (z)')
plt.ylabel('Output (a)')
plt.grid(True)

plt.tight_layout()
plt.show()

print("Derivatives (Illustrative):")

def sigmoid_derivative(x):
    return sigmoid(x) * (1 - sigmoid(x))

def relu_derivative(x):
    return np.array([1 if val > 0 else 0 for val in x])

def tanh_derivative(x):
    return 1 - np.tanh(x)**2

z_example = np.array([-1, 0, 1])
print(f"Input: {z_example}")
print(f"Sigmoid Derivative: {sigmoid_derivative(z_example)}")
print(f"ReLU Derivative: {relu_derivative(z_example)}")
print(f"Tanh Derivative: {tanh_derivative(z_example)}")
```

**Activation Function in above code:**
* **`sigmoid(x)`:** Implements the sigmoid function, which outputs values between 0 and 1.
* **`relu(x)`:** Implements the Rectified Linear Unit, which outputs the input if it's positive and 0 otherwise.
* **`tanh(x)`:** Implements the hyperbolic tangent function, which outputs values between -1 and 1.
* **`step_function(x)`:** Implements a simple step function (threshold at 0).

In the python code, we apply each activation function to the input values (`z`) and plot the resulting output (`a`) to visualize their non-linear transformations. We also define the derivatives of these activation functions. The derivatives are crucial during the backpropagation process for calculating gradients.

Note that activation functions are indispensable for enabling neural networks to learn complex, non-linear patterns in data. The choice of activation function can significantly impact the network's performance, learning speed, and the types of problems it can effectively solve.

**Attention**: Different layers in a neural network might use different activation functions depending on the task and the desired properties of the output at that stage.

## Scenario:
You are given a neural network with 10 neurons in the input layer, 3 hidden layers containing 7, 6, and 5 neurons in the first, second, and third hidden layers, respectively, and 3 neurons in the output layer.

Construct the neural network architecture diagram.

```python
import matplotlib.pyplot as plt
import networkx as nx

# Define the number of neurons in each layer
input_layer_size = 10
hidden_layer_sizes = [7, 6, 5]
output_layer_size = 3

# i. Define Depth and Width and Identify for this Network

def define_depth_width():
    """Defines depth and width of a neural network."""
    print("\n--- Depth and Width ---")
    print("Depth: The depth of a neural network refers to the number of layers with trainable parameters")
    print("       (i.e., hidden layers + output layer). It indicates the complexity of the computations the network can perform.")
    print("Width: The width of a neural network typically refers to the number of neurons in its widest layer.")
    print("      It indicates the capacity of a layer to learn and represent features.")

define_depth_width()

# Identify depth
depth = len(hidden_layer_sizes) + 1  # +1 for the output layer
print(f"The depth of this neural network is: {depth} (3 hidden layers + 1 output layer).")

# Identify width
width = max([input_layer_size] + hidden_layer_sizes + [output_layer_size])
print(f"The width of this neural network is: {width} (the maximum number of neurons in any layer, which is {width} in the input layer).")

# ii. Calculate the Total Number of Parameters

def calculate_total_parameters(input_size, hidden_sizes, output_size):
    """Calculates the total number of parameters (weights and biases) in a neural network."""
    total_parameters = 0

    # Parameters between input and first hidden layer
    total_parameters += (input_size * hidden_sizes[0]) + hidden_sizes[0]  # weights + biases

    # Parameters between hidden layers
    for i in range(len(hidden_sizes) - 1):
        total_parameters += (hidden_sizes[i] * hidden_sizes[i+1]) + hidden_sizes[i+1] # weights + biases

    # Parameters between the last hidden layer and the output layer
    total_parameters += (hidden_sizes[-1] * output_size) + output_size  # weights + biases

    return total_parameters

total_params = calculate_total_parameters(input_layer_size, hidden_layer_sizes, output_layer_size)
print(f"\n--- Total Number of Parameters ---")
print(f"The total number of parameters (weights and biases) in this network is: {total_params}.")

# iii. Construct the Corresponding Neural Network Architecture Diagram

def construct_network_diagram(input_size, hidden_sizes, output_size):
    """Constructs a diagram of the neural network architecture."""
    layer_sizes = [input_size] + hidden_sizes + [output_size]
    n_layers = len(layer_sizes)
    G = nx.DiGraph()

    # Add nodes for each neuron in each layer
    for i in range(n_layers):
        for j in range(layer_sizes[i]):
            G.add_node(f"L{i+1}_N{j+1}", layer=i)

    # Add edges representing connections between neurons
    for i in range(n_layers - 1):
        for j in range(layer_sizes[i]):
            for k in range(layer_sizes[i+1]):
                G.add_edge(f"L{i+1}_N{j+1}", f"L{i+2}_N{k+1}")

    # Position the nodes for visualization
    pos = {}
    for node in G.nodes():
        layer = G.nodes[node]['layer']
        pos[node] = (layer, - (G.degree(node) // 2) * 0.5 if G.degree(node) > 0 else 0) # Basic horizontal layering

    # Adjust vertical spacing based on layer size
    y_offsets = {}
    for i in range(n_layers):
        layer_nodes = [n for n, data in G.nodes(data=True) if data['layer'] == i]
        y_positions = np.linspace(0, -len(layer_nodes) + 1, len(layer_nodes))
        for idx, node in enumerate(layer_nodes):
            y_offsets[node] = y_positions[idx]

    for node, (x, _) in pos.items():
        pos[node] = (x, y_offsets[node])

    # Draw the graph
    plt.figure(figsize=(12, 8))
    nx.draw(G, pos, with_labels=False, node_size=500, node_color='skyblue', edge_color='gray', width=0.5)

    # Add layer labels
    layer_labels = {
        "L1": "Input Layer",
        "L2": "Hidden Layer 1",
        "L3": "Hidden Layer 2",
        "L4": "Hidden Layer 3",
        "L5": "Output Layer"
    }
    for i in range(n_layers):
        x_pos = i
        y_pos = 1  # Adjust vertical position for labels
        plt.text(x_pos, y_pos, layer_labels.get(f"L{i+1}", f"Layer {i+1}"), horizontalalignment='center', fontsize=10, fontweight='bold')

    plt.title("Neural Network Architecture\n")
    plt.axis('off')
    plt.show()

print("\n--- Neural Network Architecture Diagram ---")
construct_network_diagram(input_layer_size, hidden_layer_sizes, output_layer_size)
```
